```python
# 6) Напишіть запити, які вибирають товари за різними критеріям і їх сукупності: 

# a) категорія та ціна (в проміжку) - конструкція $and

category = 'Phone'
min_price = 500
max_price = 1000

products_in_range = products.find({
    '$and': [
        {'category': category},
        {'price': {'$gte': min_price, '$lte': max_price}}
    ]
})

print('Товари за категорією та ціною в проміжку:')
list(products_in_range)
```

```python
# b) модель чи одна чи інша - конструкція $or

model_1 = 'OnePlus 9 Pro'
model_2 = 'Galaxy S20'

products_with_model = products.find({
    '$or': [
        {'model': model_1},
        {'model': model_2}
    ]
})

print('Товари за моделлю (одна або інша):')
list(products_with_model)
```

```python
# c) виробники з переліку - конструкція $in

producers_list = ["Apple", "Samsung", "Xiaomi"]

products_with_producers = products.find({
    'producer': {'$in': producers_list}
})

print('Товари від виробників з переліку:')
list(products_with_producers)
```
